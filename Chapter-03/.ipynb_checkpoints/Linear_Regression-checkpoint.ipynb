## <font style = "color:rgb(50,120,229)">Multiple Linear Regression</font>

```r
lm.fit = lm(medv ~ lstat + age, data = Boston)
```

```r
summary(lm.fit)
```

```r
lm.fit = lm(medv~., data=Boston)
```

```r
summary(lm.fit)
```

```r
?summary.lm
```

```r
summary(lm.fit)$r.sq
```

```r
summary(lm.fit)$sigma
```

```r
library(car)
```

```r
vif(lm.fit)
```

```r
lm.fit = lm(medv~.-age,data=Boston)
```

```r
summary(lm.fit)
```

```r
lm.fit = update(lm.fit, ~.-age)
```

```r
summary(lm.fit)
```

## <font style = "color:rgb(50,120,229)">Interaction Terms</font>

```r
summary(lm(medv~lstat*age,data=Boston))
```

## <font style = "color:rgb(50,120,229)">Non-linear Transformations of the Predictors</font>

```r
lm.fit2 = lm(medv~lstat+I(lstat^2))
```

```r
summary(lm.fit2)
```

```r
lm.fit = lm(medv~lstat)
```

```r
anova(lm.fit,lm.fit2)
```

```r
par(mfrow=c(2,2))
plot(lm.fit2)
```

```r
lm.fit5 = lm(medv~poly(lstat,5))
```

```r
summary(lm.fit5)
```

```r
summary(lm(medv~log(rm),data=Boston))
```
